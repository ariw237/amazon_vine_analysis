```python
import pandas as pd 
import numpy as np
```

```python
#Read in the data
columns = ["review_id", 
           "star_rating", 
           "helpful_votes", 
           "total_votes", 
           "vine", 
           "verified_purchase"]
vine_data_df = pd.read_csv("vine_table.csv", names=columns)
```

```python
#Verify columns
vine_data_df.head()
```

```python
#Filter for total votes >= 20
vine_df_filtered_total_votes = vine_data_df.loc[(vine_data_df['total_votes'] >= 20)]
vine_df_filtered_total_votes.head()
```

```python
#Filter for helpful_votes/total_votes >= 0.50
vine_df_filtered_votes_ratio = vine_df_filtered_total_votes.loc[(
    vine_df_filtered_total_votes["helpful_votes"] >= 
    vine_df_filtered_total_votes["total_votes"]/2)]
vine_df_filtered_votes_ratio.head()
```

```python
final_df_vine = vine_df_filtered_votes_ratio.loc[(vine_df_filtered_votes_ratio["vine"]=="Y")]
final_df_vine.head()
```

```python
final_df_no_vine = vine_df_filtered_votes_ratio.loc[(vine_df_filtered_votes_ratio["vine"]=="N")]
final_df_no_vine.head()
```

```python
final_df_vine['star_rating'].value_counts()
```

```python
vine_five_star_percentage = (final_df_vine.loc[(final_df_vine['star_rating'] == 5)].
                        count().loc['star_rating'])/final_df_vine.size*100
```

```python
non_vine_five_star_percentage = (final_df_no_vine.loc[(final_df_no_vine['star_rating'] == 5)].
                        count().loc['star_rating'])/final_df_no_vine.size*100
```

```python
total_non_vine_reviews = final_df_no_vine.size
total_vine_reviews = final_df_vine.size
total_non_vine_five_star = final_df_no_vine.loc[(final_df_no_vine['star_rating'] == 5)].count().loc['star_rating']
total_vine_five_star = final_df_vine.loc[(final_df_vine['star_rating'] == 5)].count().loc['star_rating']
print("Total number of reviews for paid vine users: " + str(total_vine_reviews))
print("Total number of 5-star reviews for paid vine users: " + str(total_vine_five_star))
print("Percentage of five star reviews paid vine users: " + str(vine_five_star_percentage))
print("--------------------")
print("Total number of reviews for unpaid vine users: " + str(total_non_vine_reviews))
print("Total number of 5-star reviews for unpaid vine users: " + str(total_non_vine_five_star))
print("Percentage of five star reviews for unpaid vine users: " + str(non_vine_five_star_percentage))
```
